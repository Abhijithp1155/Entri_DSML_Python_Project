```python
import pandas as pd # Importing Panda library.
import numpy as np # Importing Numppy library.
import matplotlib.pyplot as plt # Importing matplotlib library.
```

### Importing Dataset 


```python
df = pd.read_excel('myexcel.xlsx') # Importing dataset which is in excel format.
df.head(25) # showing first 25 rows.
```

### Extracting basic informations from Data

```python
print(df.describe()) # To give some mathematical calculations of the dataset before pre-processing.
print()
print(df.info()) # To give basic information of the dataset before pre-processing.
print()
print(df.shape) # To give No of rows and columns of the dataset before pre-processing.
```

## Data Cleaning

#### Substituing blank values with Minimum Salary value in Salary Column

```python
min_value=df["Salary"].min() # To find the minimum value of salary.
df['Salary'].fillna(value=min_value,inplace=True) # To Substitute the minimum value of salary in blank columns of Salary.
```

#### Filling blank College names with Not Available

```python
College_name="Not Avaliable" # The message to be displayed in the blank columns of college name.
df['College'].fillna(College_name,inplace=True) # Substituing the blank columns with the above mentioned college name.
```

#### Changing the Height value with random number between 150 & 180

```python
df["Height"]=np.random.randint(150,180,size=len(df)) # Changing the wrong format value with random numbers between 150 & 180.
df.head() # To display first 5 rows.
```

### Basic information after pre processing data

```python
print(df.describe())  # To give some mathematical calculations of the dataset after processing.
print()
print(df.info()) # To give basic information of the dataset after pre-processing.
print("\nNo.of rows and Columns: ",df.shape)  # To give No of rows and columns of the dataset before pre-processing.
```

# QnA

## 1. How many are there in each Team and the percentage splitting with respect to the total employees.


```python
emp_team=df.groupby("Team")["Team"].count().sort_values(ascending=True) # To find no.of employees working in each team & sort values in ascending order.
print("Total number of employees in each team\n",emp_team) #To print that value.
print()

total_emp_team=df["Team"].count() # To find the total team count.
emp_team_perc=((emp_team/total_emp_team)*100).sort_values(ascending=True) # To find the percentage distribution of employees in each team.

print("Percentage splitting of each team\n",emp_team_perc) #To print that value.
```

## 2. Segregate the employees w.r.t different positions.

```python
emp_position=df.groupby("Position")["Position"].count().sort_values(ascending=True) # Grouping total no of employees with respect to differnt positions.
print("Segregating employees w.r.t different position \n",emp_position) #To print that value.

label=df['Position'].value_counts().index # To get the label as Position name.

plt.figure(1,figsize=(7,7)) # To fix the size of the chart to be drawn.
plt.pie(emp_position, labels=label,shadow=True) # To draw the values in a pie chart format.
plt.legend() # To show the labels with color.
plt.show() # To show the pie chart.
```

## 3. Find from which age group most of the employees belong to.

```python
emp_count=df.groupby("Age")["Age"].count().sort_values(ascending=False) # To find the age count.
age_group=df.groupby("Age")["Age"].count().sort_values(ascending=False).index # Belongs to which age group.

#Visualizing which age group employees are working mostly in this ABC Company.

plt.figure(figsize=(10,12)) # To fix the size of the bar chart.
plt.bar(age_group,emp_count) #To plot bar chart.
plt.xticks(age_group) # To show all x-labels.
plt.yticks(emp_count) # To show all y-labels.
plt.xlabel("Age group") # To give label for x axis.
plt.ylabel("No.of Employees") # To give label for y axis.

plt.title("Age Vs No.of employees") # To give Title for bar graph.
plt.show() # To show the bargraph.
```

```python
print("The above graph clearly shows, most of the employees working in ABC Company having an age of \n\n",emp_count[0:2])
```

## 4. Find out under which team and position, spending in terms of salary is high.

```python
high_sal=df.groupby(["Team","Position"])["Salary"].sum().sort_values(ascending=False) # To find which team and position need more salary

print("Below mentioned team and position is the top three, spending high in terms of salary in ABC Company \n\n",high_sal[0:3])
```

## 5.Find if there is any correlation between age and salary , represent it visually.

```python
correlation=df["Age"].corr(df["Salary"]) # To find the correlation between Age & Salary

plt.scatter(df["Age"],df["Salary"],c="r") # To vizualize the correlation of age and salary in scatter plot format
plt.xlabel("Age") # To give name to the x axis
plt.ylabel("Salary") # To give name to the y axis
plt.title("Correlation: "+str(correlation)) # To give title to the Scatter plot along with the correlation value
plt.show() # To show the graph 
```

```python
print("This shows that there is a weak positive correlation between Age & Salary")
```
